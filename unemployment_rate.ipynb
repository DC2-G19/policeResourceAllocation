```python
import pandas as pd

read_file = pd.read_excel (r'unemployment_rates_part.xlsx')
read_file.to_csv (r'unemployment_rates_part.csv', index = None, header=True)
```

```python
df = pd.read_csv('unemployment_rates_part.csv')
```

```python
df
```

```python
london_unempl_rates = df.iloc[[0]]
london_unempl_rates = london_unempl_rates.drop('Quarter ending', axis=1)
london_unempl_rates
```

```python
df_transposed = london_unempl_rates.T
df_transposed.index = pd.to_datetime(df_transposed.index).strftime('%Y-%m')
df_transposed = df_transposed.rename(columns={0: 'Unemployment Rate'})

df_transposed
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))

plt.plot(df_transposed.index, df_transposed)

plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Values for Row1')
plt.xticks(rotation=90)

plt.show()
```

```python
orig_data = pd.read_csv('conc_clean.csv')
orig_data_burgl = orig_data[orig_data['Crime type']=='Burglary']
orig_data_burgl
```

```python
crimes_per_month = orig_data_burgl.groupby('Month').size()
df_crimes = pd.DataFrame({'Month': crimes_per_month.index, 'Number of Crimes': crimes_per_month.values})
df_crimes.set_index('Month', inplace=True)

df_crimes
```

```python
fig, ax1 = plt.subplots(figsize=(20, 10))
df_crimes.plot(kind='line', ax=ax1)

ax1.set_title('Group Counts')
ax1.set_xlabel('Date')
ax1.set_ylabel('Burglary Count')

df_transposed.index = pd.to_datetime(df_transposed.index).strftime('%Y-%m')
ax2 = ax1.twinx()  # Create a twin Axes sharing the x-axis
ax2.plot(df_transposed.index, df_transposed, color='red')
ax2.set_ylabel('London Unemployment Rate', color='red')

# Reduce the number of dates displayed on x-axis of ax2
tick_locations = ax2.get_xticks()
subset_ticks = tick_locations[::3]  # Display every other tick for ::2

ax2.set_xticks(subset_ticks)
ax1.xaxis.set_tick_params(rotation=90)

plt.show()
```

```python
import statsmodels.api as sm

## Convert the columns to numeric data types
df_crimes["Number of Crimes"] = pd.to_numeric(df_crimes["Number of Crimes"])
df_transposed['Unemployment Rate'] = pd.to_numeric(df_transposed['Unemployment Rate'])

# make sure the indices are the same
df_combined = pd.concat([df_crimes, df_transposed], axis=1, join='inner')
y = df_combined["Number of Crimes"]
x = df_combined['Unemployment Rate']

model = sm.OLS(y, x).fit()
print(model.summary())

# coefficient of 912.8959, R2=0.949, so good dependence unemployment rate and nr burglaries
```

```python
plt.scatter(df_combined['Unemployment Rate'], df_combined['Number of Crimes'])

plt.title('Scatter Plot of Unemployment Rate and Number of Crimes')
plt.xlabel('Unemployment Rate')
plt.ylabel('Number of Crimes')

plt.show()

# seems a positive correlation in groups
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
kmeans.fit(df_combined)

labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_
num_clusters = len(cluster_centers)
colors = ['red', 'blue', 'green', 'orange', 'purple', 'yellow']

# Plot the data points with different colors for each cluster
for i in range(len(df_combined)):
    cluster_label = labels[i]
    plt.scatter(df_combined.iloc[i, 0], df_combined.iloc[i, 1], c=colors[cluster_label])

# Plot the cluster centers
for i in range(num_clusters):
    plt.scatter(cluster_centers[i, 0], cluster_centers[i, 1], c='black', marker='x')

plt.xlabel('Number of Crimes')
plt.ylabel('Unemployment Rate')
plt.title('K-means Clustering')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Assuming your DataFrame is named 'df_monthly' with 'Unemployment Rate' column

# Convert the 'Unemployment Rate' column to numeric if it's not already
df_monthly['Unemployment Rate'] = pd.to_numeric(df_monthly['Unemployment Rate'], errors='coerce')

# Perform seasonal decomposition
result = seasonal_decompose(df_monthly['Unemployment Rate'], model='additive', period=12)

# Plot the decomposition components
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(10, 8))
result.observed.plot(ax=ax1)
ax1.set_ylabel('Observed')
result.trend.plot(ax=ax2)
ax2.set_ylabel('Trend')
result.seasonal.plot(ax=ax3)
ax3.set_ylabel('Seasonal')
result.resid.plot(ax=ax4)
ax4.set_ylabel('Residual')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```
