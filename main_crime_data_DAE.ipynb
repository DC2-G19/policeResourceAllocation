```python

# Create a new figure and axis
fig, ax = plt.subplots(figsize=(15, 8))

# Plot the data
sns.lineplot(data=combined_df_copy, x='Month', y='Number of Burglaries', ax=ax, color='black', linewidth='3', marker='o')

# Define the seasons and their corresponding colors
seasons = {
    'Spring': ('03-01', '05-31', 'green'),
    'Summer': ('06-01', '08-31', 'orange'),
    'Fall': ('09-01', '11-30', 'red'),
    'Winter': ('12-01', '02-28', 'blue')
}

# Add the seasonal background colors
for year in range(2010, 2023):
    for season, (start, end, color) in seasons.items():
        if season == 'Winter':
            ax.axvspan(pd.Timestamp(f'{year}-{start}'), pd.Timestamp(f'{year + 1}-{end}'),
                       alpha=0.1, color=color)
        else:
            ax.axvspan(pd.Timestamp(f'{year}-{start}'), pd.Timestamp(f'{year}-{end}'),
                       alpha=0.1, color=color)

ax.set_xlim(pd.Timestamp('2010-12-01'), pd.Timestamp('2023-01-01'))

# Create legend patches for each season
legend_patches = [mpatches.Patch(color=color, label=season, alpha=0.3) for season, (_, _, color) in seasons.items()]

# Add legend to the plot
ax.legend(handles=legend_patches, loc='upper right', bbox_to_anchor=(0.98, 0.95), prop={'size': 17})

# Set the plot title and axis labels
plt.title('Number of Burglaries by Month in Barnet', fontsize=17)
plt.xlabel('Year (Seasons color encoded)', fontsize=13)
plt.ylabel('Number of Burglaries', fontsize=13)

# Show the plot
plt.show()
```

```python


# Group the dataframe by Crime type
crime_type_groups = df_burglary_all.groupby('Crime type')

# Create separate dataframes for each crime type
crime_type_dfs = {}
for crime_type, group in crime_type_groups:
    crime_type_dfs[crime_type] = group.copy()
    
```

```python
count_dict = {}
pivot_dict = {}

for crime_type, df in crime_type_dfs.items():
    count_df = df.groupby(['Month', 'Crime type']).size().reset_index(name='Count')
    pivot_df = count_df.pivot_table(values='Count', index='Month', columns='Crime type')
    
    count_dict[crime_type] = count_df
    pivot_dict[crime_type] = pivot_df
```

```python
for key, value in pivot_dict.items():
    print(key)
```

```python
import matplotlib.pyplot as plt

# create a new figure
fig, ax = plt.subplots(figsize=(10, 6))

# iterate over the dataframes in pivot_dict and plot each one
for crime_type, df in pivot_dict.items():
    ax.plot(df.index, df.values, label=crime_type)

# set the x-axis label
ax.set_xlabel('Month')

# set the y-axis label
ax.set_ylabel('Count')

# set the plot title
ax.set_title('Crime counts by month')

# add a legend
ax.legend(fontsize=6)

# show the plot
plt.show()
```
