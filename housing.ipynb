```python
import pandas as pd
import os
```

```python
#housing DF
housing_df_2c = pd.read_excel('Data\Copy of hou01tables2021.xlsx', sheet_name=['2c'])['2c']
housing_df_2c = housing_df_2c.replace('c', 0)
```

```python
housing_df_3c = pd.read_excel('Data\Copy of hou01tables2021.xlsx', sheet_name=['2c'])['2c']
housing_df_3c = housing_df_2c.replace('c', 0)
```

```python
housing_df_2c['Proportion of indep']=(housing_df_2c['Detached whole house or bungalow']+housing_df_2c['Semi-detached whole house or bungalow']+housing_df_2c['Terraced (including end-terrace) whole house or bungalow'])/housing_df_2c['Flat, maisonette or apartment']
#adding col for ratio
housing_df_2c=housing_df_2c.rename(columns={'Area Code': 'LSOA code'})
housing_df_2c.head()
```

```python
housing_df_3c = pd.read_excel('Data\Copy of hou01tables2021.xlsx', sheet_name=['3c'])['3c']
housing_df_3c = housing_df_3c.replace('c', 0)
housing_df_3c=housing_df_3c.rename(columns={'Area Code': 'LSOA code'})
housing_df_3c['Proportion social']=housing_df_3c['Social Rented']/(housing_df_3c['Social Rented']+housing_df_3c['Private Rented or lives rent free']+housing_df_3c['Owned or shared ownership'])
housing_df_3c['Proportion rented']=(housing_df_3c['Social Rented']+housing_df_3c['Private Rented or lives rent free'])/(housing_df_3c['Social Rented']+housing_df_3c['Private Rented or lives rent free']+housing_df_3c['Owned or shared ownership'])
housing_df_3c
```

```python
housing_df_4c = pd.read_excel('Data\Copy of hou01tables2021.xlsx', sheet_name=['4c'])['4c']
housing_df_4c = housing_df_4c.replace('c', 0)
housing_df_4c=housing_df_4c.rename(columns={'Area Code': 'LSOA code'})
housing_df_4c
```

```python
housing_df_7c = pd.read_excel('Data\Copy of hou01tables2021.xlsx', sheet_name=['7c'])['7c']
housing_df_7c = housing_df_7c.replace('c', 0)
housing_df_7c=housing_df_7c.rename(columns={'Area Code': 'LSOA code'})
housing_df_7c['Proportion small house']=(housing_df_7c['1 bedroom']+housing_df_7c['2 bedrooms'])/(housing_df_7c['1 bedroom']+housing_df_7c['2 bedrooms']+housing_df_7c['3 bedrooms']+housing_df_7c['4 or more bedrooms'])
housing_df_7c
```

```python
merged_df_temp = pd.merge(housing_df_3c, housing_df_2c, on=['LSOA code', 'Area Name'], how='outer')
merged_df = pd.merge(merged_df_temp, housing_df_7c, on=['LSOA code', 'Area Name'], how='outer')
merged_df=merged_df.rename({"Area Name":"LSOA name"}, axis=1)
merged_df
```

```python
merged_df.to_excel('Data/housing_combined.xlsx', index=False)
```
